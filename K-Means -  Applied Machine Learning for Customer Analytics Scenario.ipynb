### Using *k*-means for customer segmentation



### Demo 1

1. Importing the libraries
2. Importing the dataset
3. Training the k-means model on the dataset with k=2
4. Visualizing the clusters

#### 1. Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='retina' # display figures in retina quality

from sklearn.cluster import KMeans
```

#### 2. Importing the dataset

```python
df = pd.read_csv('customers.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.hist(df['Annual_Income'], bins=30, color = 'green', edgecolor='black')
plt.title('Annual Income')
plt.show()
```

```python
plt.hist(df['Spending_Score'], bins=30, color = 'blue', edgecolor='black')
plt.title('Spending Score')
plt.show()
```

#### 3. Training the k-means model on the dataset with k=2

```python
# k ==2 selected as inital example, optimal k value calculated below uisng silhouette score 

# 'k-means++': selecting initial cluster centroids in a smart way to speed up convergence

# random_state: determining random number generation for centroid initialization
#               using an integer to make the randomness deterministic


kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 337)
kmeans
```

```python
# selecting only the 2nd column (annual income) and the 3rd column (spending score) from the dataset

# df.values: change dataframe to numpy array

X = df.iloc[:, [1,2]].values

# .values removes the titles and only uses the vlaues as a numpy array

type(X)
```

```python
# using data to fit the model and compute k-means clustering

kmeans.fit(X)
```

```python
# assigning each data point to the closest cluster

y_kmeans = kmeans.predict(X)
```

```python
y_kmeans
```

#### 4. Visualizing the clusters

```python
plt.scatter(X[y_kmeans==0, 0], X[y_kmeans==0, 1], c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans==1, 0], X[y_kmeans==1, 1], c = 'blue', label = 'Cluster 2')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 500, c = 'orange', marker = "*")

plt.title('Customer Segmentation')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')

plt.legend()
plt.show()
```

### Demo 2: The Elbow Method

1. Importing the libraries
2. Importing the dataset
3. Using the __*elbow method*__ to find the optimal number of clusters
3. Training the k-means model on the dataset with the optimal number of clusters
4. Visualizing the clusters

```python
RSS = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 337)
    kmeans.fit(X)
    RSS.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), RSS, c='orange')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('RSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 337)
kmeans.fit(X)
```

```python
y_kmeans = kmeans.predict(X)
y_kmeans
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], c = 'magenta', label = 'Cluster 5')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 500, c = 'orange', marker = "*")

plt.title('Customer Segmentation')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')

plt.legend()
plt.show()
```

### Demo 3: Silhouette Score

1. Importing the libraries
2. Importing the dataset
3. Using the __*silhouette score*__ to find the optimal number of clusters
3. Training the k-means model on the dataset with the optimal number of clusters
4. Visualizing the clusters

```python
from sklearn import metrics
```

```python
best_score = -np.inf     # initialize with -1 or lower (bottom end of sil. score range. Don't use 0.)
best_k = None

for k in range(2, 11): #start at i = 2 so there is cross cluster variation
    candidate_kmeans = KMeans(n_clusters = k, init = 'k-means++', random_state = 337)
    candidate_kmeans.fit(X)
    cluster_assignments = candidate_kmeans.predict(X)
    
    score = metrics.silhouette_score(X, cluster_assignments) # calc. sil. score using initial data and assigned clusters
    
    print('k =', k, 'has silhouette score', score)
    
    if score > best_score:   # update the values using the accumulator pattern
        best_score = score
        best_k = k
    
print('\nBest k according to silhouette score:', best_k)
print('Best silhouette score:', best_score)
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 337)
kmeans.fit(X)
```

```python
y_kmeans = kmeans.predict(X)
y_kmeans
```

```python
# grahing the cluster with k-value with the best sil. score (5)  

plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], c = 'magenta', label = 'Cluster 5')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 500, c = 'orange', marker = "*")

plt.title('Customer Segmentation')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')

plt.legend()
plt.show()
```
